```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/greendestination (1).csv')
```

```python
df.head()
```

```python
#counting the number of rows
row_count = len(df)
print(row_count)
```

### ***Attrition Rate***

```python
attrition_rate = (df['Attrition'] == 'Yes').mean() * 100
print("Attrition Rate:", attrition_rate, "%")
```

### ***Dependency of age on attrition***

```python
plt.figure(figsize=(15, 9))
sns.countplot(x='Age', hue='Attrition', data=df)
plt.title('Attrition by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# From the above graph we can conclude that the attrition rate reduces as their age or experience increases in the organisation. Young ones or freshers have a higher rate of attrition as compared to experienced employees.

# ***Dependency of age on attrition***

```python
plt.figure(figsize=(15, 9))
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Years at Company vs. Attrition')
plt.xlabel('Attrition')
plt.ylabel('Years at Company')
plt.show()
```

```python
plt.figure(figsize=(15, 9))
sns.swarmplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Years at Company vs. Attrition')
plt.xlabel('Attrition')
plt.ylabel('Years at Company')
plt.show()
```
